```python
A = np.array([[3, 7], [5, 2]])
print(A)
```

```python
U, D, V = np.linalg.svd(A)

print(D[0])
print(D[1])
```

```python
u1 = [D[0] * U[0,0], D[0] * U[0,1]]
v1 = [D[1] * U[1,0], D[1] * U[1,1]]

print([A[0,0], A[1,0]])
print(u1)
print()
print([A[0,1], A[1,1]])
print(v1)
```
